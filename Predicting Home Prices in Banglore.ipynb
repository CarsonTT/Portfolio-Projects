```python
# Importing libraries


import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
# Importing dataset


df1 = pd.read_csv("D:\Downloads\Spreedsheets\Bengaluru_House_Data.csv")
df1
```

```python
# Dropping unwanted columns


df2 = df1.drop(['area_type', 'society', 'balcony', 'availability'], axis= 'columns')
df2
```

```python
# Checking for null values


df2.isnull().sum()
```

```python
# Dropping null values


df3 = df2.dropna()
df3.isnull().sum()
```

```python
# Cleaning column "size" p1


df3['size'].unique()
```

```python
# Cleaning column, "size" p2


df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))

df3
```

```python
# Cleaning column, "total_sqft" p1


# Checking values for total _sqft

df3[df3.bhk>20]

# NOTES
# 43 bedrooms cant be equal to 2400 sqft
```

```python
# Cleaning column, "total_sqft" p2


def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

df3[~df3['total_sqft'].apply(is_float)]
```

```python
# Cleaning column, "total_sqft" p3


def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None   
    

# Testing function
    
convert_sqft_to_num('2166')
```

```python
convert_sqft_to_num('2100 - 2850')
```

```python
convert_sqft_to_num('34.46Sq. Meter')
```

```python
# Cleaning column, "total_sqft" p4 


df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)

df4
```

```python
# Finding price per square feet


df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']

df5
```

```python
# Cleaning column, "location" p1


len(df5.location.unique())
```

```python
# Cleaning column, "location" p2



# Finding how many values in a category

df5.location = df5.location.apply(lambda x: x.strip())

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)

location_stats.head(20)
```

```python
# Cleaning column, "location" p2


len(location_stats[location_stats<10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]

location_stats_less_than_10
```

```python
len(df5.location.unique())
```

```python
# Cleaning column, "location" p3


df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)

len(df5.location.unique())
```

```python
# Finding rows that dont seem to have the right total sqft compared to the # of bedrooms


df5[df5['total_sqft']/df5['bhk']<300]
```

```python
# Removing inaccurate data


df6 = df5[~(df5['total_sqft']/df5['bhk']<300)]

df6
```

```python
df6.price_per_sqft.describe()
```

```python
# Removing outliers


def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df7 = remove_pps_outliers(df6)

df7
```

```python
# removing apts with higher bedrooms that cost less than apts with lower bedrooms


def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df8 = remove_bhk_outliers(df7)

df8
```

```python
# Removing apts that have 3+ bathrooms compared to bedrooms


df9 = df8[df8.bath<df8.bhk+2]

df9
```

```python
# Dropping unwanted columns


df10 = df9.drop(['size','price_per_sqft'], axis = 'columns')

df10
```

```python
# Creating dummies


dummies = pd.get_dummies(df10.location)
dummies
```

```python
df11 = pd.concat([df10, dummies.drop('other', axis = 'columns')], axis = 'columns')

df11
```

```python
df12 = df11.drop('location', axis = 'columns')

df12.head(1)
```

```python
# Making model


x = df12.drop('price', axis = 'columns')

y = df12.price


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)


from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
# Finding best model


from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```

```python
# making dunciton to rpedict price based on model


def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(x.columns==location)[0][0]

    X = np.zeros(len(x.columns))
    X[0] = sqft
    X[1] = bath
    X[2] = bhk
    if loc_index >= 0:
        X[loc_index] = 1

    return lr_clf.predict([X])[0]
```
